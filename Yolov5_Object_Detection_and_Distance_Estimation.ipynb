#Semi Supervised Annotation

```python
import torch
import cv2    
import os
model = torch.hub.load('/content/yolov5/', 'custom', path='/content/yolov5/runs/train/exp3/weights/best.pt', source='local') 
annotation_path = "/content/Annotations"
image_path = "/content/drive/MyDrive/Information/Phase2/SemiSupervised Training/train/"
image_list = os.listdir(image_path)
print(image_list)


def bbox_generator(annotation_path,image):  
  img = cv2.imread(image)
  h,w,_ = img.shape
  #print(w,h)
  results = model(image)
  coords = results.pandas().xyxy[0]
  #print(coords)
  
  for index, row in coords.iterrows():
    x_center = str(((row["xmin"] + row["xmax"])/2)/w)
    y_center = str(((row["ymin"] + row["ymax"])/2)/h)
    x_width = str((row["xmax"] - row["xmin"])/w)
    y_height = str((row["ymax"] - row["ymin"])/h)
    predicted_class = str(row["class"])
    
    line = " ".join((predicted_class, x_center, y_center, x_width, y_height, "\n"))
    text_file_name = str(image).replace(".jpg", ".txt")
    file_location = os.path.join(annotation_path, text_file_name)
    
    with open(file_location,'a') as text_file:
      text_file.write(line)

  #return x_center, y_center, x_width, y_height

for image_name in image_list:
  image = image_path + image_name
  bbox_generator(annotation_path, image)
  print("Done", image_name)
#print(xc, yc, xw, yh)
```

## Focal Length Calculation

```python
import torch
import cv2    
import os
model = torch.hub.load('/content/yolov5/', 'custom', path='/content/yolov5/runs/train/exp4/weights/best.pt', source='local') 

initial_distance = 52
product_widths_real = [12.67, 5.416, 9.067]

safekeeper_widths = []
spa_widths = []
superstar_widths = []

def focal_length_finder(image):  
  
  results = model(image)
  coords = results.pandas().xyxy[0]
  
  for index, row in coords.iterrows():
    if row['class'] == 0:
      safekeeper_widths.append(row['xmax'] - row['xmin'])
    elif row['class'] == 1:
      spa_widths.append(row['xmax'] - row['xmin'])
    elif row['class'] == 2:
      superstar_widths.append(row['xmax'] - row['xmin'])

image_list = os.listdir("/content/SemiSuperXXC-1/train/images/") 

for img in image_list:
  image_name = "/content/SemiSuperXXC-1/train/images/" + str(img)
  focal_length_finder(image_name)
try:
  avg_safekeeper_px = sum(safekeeper_widths)/len(safekeeper_widths)
  #print(len(safekeeper_widths))
  avg_spa_px = sum(spa_widths)/len(spa_widths)
  #print(len(spa_widths))
  avg_superstar_px = sum(superstar_widths)/len(superstar_widths)
  #print(len(superstar_widths))
except:
  print("error")



safekeeper_focal_length = (avg_safekeeper_px * initial_distance)/product_widths_real[0]
spa_focal_length  = (avg_spa_px * initial_distance)/product_widths_real[1]
superstar_focal_length = (avg_superstar_px * initial_distance)/product_widths_real[2]

print(safekeeper_focal_length, spa_focal_length, superstar_focal_length)

#labels, cord_thres = results.xyxyn[0][:, -1].cpu().numpy(), results.xyxyn[0][:, :-1].cpu().numpy()
```
